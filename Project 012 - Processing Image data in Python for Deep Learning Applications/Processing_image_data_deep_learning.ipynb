 ## Resizing the Image  and saving the img
 ### (using pillow library)



```python
#Resize the image

from PIL import Image
dimg = Image.open('dog.jpg')
dimg_resized= dimg.resize((600,400))
cimg = Image.open('cat.jpg')
cimg_resized= cimg.resize((600,400))
```

```python
dimg_resized.save('res_dog.jpg')
cimg_resized.save('res_cat.jpg')
```

```python
dimg = mpimg.imread('res_dog.jpg')
img_plot=plt.imshow(dimg)
plt.show()
```

```python
cimg = mpimg.imread('res_cat.jpg')
img_plot=plt.imshow(cimg)
plt.show()
```

```python
print(cimg.shape,dimg.shape)
```

### Converting RGB image to Graysclae Image using OpenCV

```python
import cv2
```

```python
dimg=cv2.imread('res_dog.jpg')
cimg=cv2.imread('res_cat.jpg')
```

```python
type(dimg)
```

```python
print(cimg.shape,dimg.shape)
```

# Gray-scale image

```python
gs_dimg = cv2.cvtColor(dimg,cv2.COLOR_RGB2GRAY)
gs_cimg = cv2.cvtColor(cimg,cv2.COLOR_RGB2GRAY)
```

```python
type(gs_dimg)
```

```python
print(gs_dimg.shape,gs_cimg.shape)
```

## saving Grayscale Image using cv2

```python
# saving Grayscale Image using cv2

cv2.imwrite('grayscale_res_dog.jpg', gs_dimg)
cv2.imwrite('grayscale_res_cat.jpg', gs_cimg)
```
