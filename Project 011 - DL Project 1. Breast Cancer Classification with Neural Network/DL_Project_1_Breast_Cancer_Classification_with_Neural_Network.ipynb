**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

0 --> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Standardizing the Data

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
#X = pd.DataFrame(X.values)
X_train_std = scaler.fit_transform(X_train) # you need to fit data to train data only
X_test_std = scaler.transform(X_test) #
print(X_test_std)
```

# Building the Neural Network

No of Inputs = No of Features

Neuron is fired while another neuron is inactive in case of Output layer neurons .

Activation Method for Deep Learning NN

Deep Learning Course Check

download.png

```python
# Importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #inputs
    keras.layers.Dense(20 , activation='relu'), # Hidden Layer - activation='relu' rectified linear units
    keras.layers.Dense(30 , activation='sigmoid'),
    keras.layers.Dense(2 , activation='sigmoid')
]) # ouput Layer - Sigmoid function
    # ouput Layer - 2 cause there is 2 class (0 and 1 ; 1 --> Benign ; 0 --> Malignant )
```

```python
#compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
```

dog - 0    [1 0 0]

cat- 1     [0 1 0]  

horse - 2  [0 0 1]

One Hot Encoding

```python
# training the network


history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10) #epochs is how many times data will go through the model
```

# Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])



plt.title('Model Accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.legend(['train_data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])



plt.title('Model Loss')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.legend(['train_data','validation_data'],loc='upper right')
```

# Getting Accuracy value for test dataset

```python
loss, accuracy = model.evaluate(X_test_std,Y_test)
print(loss,accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
Y_pred=model.predict(X_test_std)
print(Y_pred[0])
### In this case Model predicts that probability of  the data point being 0 is 0.25196776 and 1 is  0.45538756
```
